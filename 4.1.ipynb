## Unit-4 : Regression - Model Training and Evaluation

- Linear Regression is a statistical method used to model relation b/w a dependent variable and one or more independent variables
- It finds the best fitting line which describes the relation b/w 2 variables
- Dependent variable (y) : predict
- Independent variable (x) : predictors or variables we use to predict the dependent var
- y=mx+c

- MAE : mean absolute error (average of absolute)
- MSE : mean squared error (mean of square)
- R^2 Score : 1-(SSres/SStotal)    (SSres): Sum of square - residual/error   (SStotal): Sum of square - total

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
y_pred=lr.predict(x)
print(y_pred)
```
