```python
import pandas as pd
```

```python
a1=pd.DataFrame({
    "id":[1,2,3],
    "name":["amit","riya","john"],
    "salary":[50000,60000,55000]
})

a2=pd.DataFrame({
    "id":[1,2,4],
    "name":["amit","riya","david"],
    "salary":[50000,60000,58000]
})

r=pd.merge(a1,a2,on="id",suffixes=("_left","_right"))
print(r)
```

```python
b1=pd.DataFrame({"student":[1,2,3],
                "name":["amit","ria","john"]})
b2=pd.DataFrame({"id":[1,2,4],
                "marks":[80,90,75]})
r1=pd.merge(b1,b2,left_on="student",right_on="id")
print(r1)
```

```python
c1=pd.DataFrame({"physics":[70,80,90]},index=[101,102,103])
c2=pd.DataFrame({"name":["alice","bob","charlie"]},index=[101,102,103])
r2=pd.merge(c1,c2,left_index=True,right_index=True)
print(r2)
```

```python
d1=pd.DataFrame({"student":[1,2,3],
                "name":["amit","ria","john"]})
d2=pd.DataFrame({"score":[70,80,90]},index=[1,2,3])
r3=pd.merge(d1,d2,left_on="student",right_index=True)
print(r3)
```

## groupby

```python
data={"product":["electronics","furniture","electronics","furniture"],
     "sales":[200,150,300,100]}
df=pd.DataFrame(data)
df
```

```python
r=df.groupby("product")["sales"].sum()
print(r)
```

```python
r=df.groupby("product")["sales"].mean()
print(r)
```

```python
r=df.groupby("product")["sales"].median()
print(r)
```

```python
r=df.groupby("product")["sales"].describe()
print(r)
```

```python
r=df.groupby("product")["sales"].sample()
print(r)
```

```python
r=df.groupby("product")["sales"].size()
print(r)
```

```python
r=df.groupby("product")["sales"].groups
print(r)
```

```python
r=df.groupby("product")["sales"].nunique()
print(r)
```

```python
r=df.groupby("product")["sales"].nth(0)
print(r)
```

```python
r=df.groupby("product")["sales"].nth(1)
print(r)
```

```python
r=df.groupby("product",axis=0,as_index=True,sort=True)["sales"].count()
print(r)
```

```python
r=df.groupby("product",axis=0,as_index=False,sort=True)["sales"].count()
print(r)
```

```python
e1=pd.DataFrame({"region":["east","east","west","west","east","west"],
               "sales":[100,150,200,120,110,130],
               "manager":["a","b","c","d","e","f"]})
e1
```

```python
def sales_range(group):
    return group["sales"].max()-group["sales"].min()     #max-min
sales=e1.groupby("region").apply(sales_range)
print(sales)
```

```python
e2=pd.DataFrame({"dept":['sales','sales','hr','hr','it','it'],
                "salary":[50000,60000,40000,45000,70000,75000],
                "bonus":[5000,6000,2000,2500,8000,8500]})
r5=e2.groupby("dept").agg('mean')
print(r5)
```

```python
e2=pd.DataFrame({"dept":['sales','sales','hr','hr','it','it'],
                "salary":[50000,60000,40000,45000,70000,75000],
                "bonus":[5000,6000,2000,2500,8000,8500]})
r5=e2.groupby("dept").agg(['mean','min','max'])
print(r5)
```

```python
e2=pd.DataFrame({"dept":['sales','sales','hr','hr','it','it'],
                "salary":[50000,60000,40000,45000,70000,75000],
                "bonus":[5000,6000,2000,2500,8000,8500]})
r5=e2.groupby("dept").agg({'salary':['min','max'],
                           'bonus':'sum'})
print(r5)
```

```python
e2=pd.DataFrame({"dept":['sales','sales','hr','hr','it','it'],
                "salary":[50000,60000,40000,45000,70000,75000],
                "bonus":[5000,6000,2000,2500,8000,8500]})
r5=e2.groupby("dept").agg(meansalary=('salary','mean'),
                          maxsalary=('salary','max'),
                         totalbonus=('bonus','sum'))
print(r5)
```
