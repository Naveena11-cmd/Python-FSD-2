### geocoding api for lat and lon

```python
def get_coords():
    api_key='96e6a25a7a7961eb790c459a27a96c9e'
    city=input('Enter city name : ')
    request_url=f'http://api.openweathermap.org/geo/1.0/direct?q={city}&limit=5&appid={api_key}'
    result=requests.get(request_url)
    c=result.json()
    print(json.dumps(c,indent=5))
    latitude=c[0]['lat']
    longitude=c[0]['lon']
    return latitude,longitude,city
lat,lon,city=get_coords()
print(city)
print("latitude",lat)
print("longitude",lon)
```

### Current Weather

```python
def current_weather():
    api_key='96e6a25a7a7961eb790c459a27a96c9e'
    request_url=f'https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}'
    result=requests.get(request_url)
    w=result.json()
    print(json.dumps(w,indent=5))
    descript=w['weather'][0]['description']
    temp=w['main']['temp']
    pressure=w['main']['pressure']
    humidity=w['main']['humidity']
    windspeed=w['wind']['speed']
    visibility=w['visibility']
    print()
    print("weather description : ",descript)
    print("temperature : ",temp)
    print("pressure : ",pressure)
    print("humidity : ",humidity)
    print("wind speed : ",windspeed)
    print("visibility : ",visibility)
current_weather()
```

### Air pollution

```python
def ap():
    api_key='96e6a25a7a7961eb790c459a27a96c9e'
    request_url=f'http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={api_key}'
    result=requests.get(request_url)
    a=result.json()
    print(json.dumps(a,indent=5))
    quality_index={1:'good',2:'fair',3:'moderate',4:'poor',5:'very poor'}
    aqi=a['list'][0]['main']['aqi']
    print("AQI : ",aqi)
    print(quality_index[aqi])
ap()
```
