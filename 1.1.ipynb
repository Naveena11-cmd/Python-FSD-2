## pandas series
it is like a column in a table (1-D array holding data of any type)

```python
import pandas as pd
```

```python
a=[1,7,2]
x=pd.Series(a)
print(x)
print(type(x))
```

```python
print(x[1])
```

change name of index

```python
a=[1,7,2]
x=pd.Series(a, index=["x","y","z"])
print(x)
print(type(x))
```

```python
print(x[1])
```

```python
print(x["y"])
```

```python
b=(1.4,2,3,4)
y=pd.Series(b)
print(y)
print(type(y))
```

```python
name=["ram","shyam","radha","geeta","seeta"]
marks=[50,60,70,80,90]
z=pd.Series(marks,index=name, name='student result')
print(z)
```

```python
calories={'day1':400, 'day2':380, 'day3':430, 'day4':390}
c=pd.Series(calories)
print(c)
```

```python
calories={'day1':400, 'day2':380, 'day3':430, 'day4':390}
c=pd.Series(calories, index=['day1','day2'])
print(c)
```

```python
calories={'day1':400, 'day2':380, 'day3':430, 'day4':390}
c=pd.Series(calories, name="calories burn per day")
print(c)
```
