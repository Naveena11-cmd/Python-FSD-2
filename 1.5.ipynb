## Drop and remove duplicates

```python
data={"A":["teamA","teamB","teamB","teamC","teamA"],
     "B":[50,40,40,30,50],
     "C":[True,False,False,False,True]}
df=pd.DataFrame(data)
df
```

```python
df.duplicated()
```

```python
df.duplicated(keep="first")
```

```python
df.duplicated(keep="last")  #checks from the last
```

```python
 df.duplicated(keep=False)  #gives true in all duplicates
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(keep="last")
```

```python
df.drop_duplicates(keep=False)
```

```python
df.drop_duplicates(keep="first",inplace=True)
```

### apply()

```python
s=pd.Series([35,67,90,45])
def pass_fail(x):
    return "pass" if x>=50 else "Fail"

result=s.apply(pass_fail)
print(result)
```

```python
df2=pd.DataFrame({
    "phy":[60,70,80],
    "chem":[65,75,85],
    "maths":[70,80,90]
})

result=df2.apply(sum)
print(result)
```

```python
df2=pd.DataFrame({
    "phy":[60,70,80],
    "chem":[65,75,85],
    "maths":[70,80,90]
})

result=df2.apply(sum,axis=1)
print(result)
```

```python
df2=pd.DataFrame({
    "phy":[60,70,80],
    "chem":[65,75,85],
    "maths":[70,80,90]
})

result=df2.apply(lambda x:x.mean())
print(result)
```

```python
df2
```

```python
df2["total"]=df2.apply(lambda x:x.sum(),axis=1)
df2
```

```python
def grade(total):
    if total>=250:
        return "A"
    elif total>=200:
        return "B"
    else:
        return "C"
```

```python
df2["grade"]=df2["total"].apply(grade)
```

```python
df2
```

```python
d=pd.DataFrame({"width":[1,2], "height":[3,4]})
def get_stats(row):
    return [row['width']*row['height'], 2*(row['width']+row['height'])]
d
```

```python
d.apply(get_stats, axis=1)  #output same as result_type="reduce"
```

```python
d.apply(get_stats, axis=1, result_type="expand")
```

```python
d.apply(get_stats, axis=1, result_type="broadcast")
```

##  args & kwargs

```python
def add_values(x,y,z):
    return x+y+z

d=pd.DataFrame({"A":[1,2],"B":[10,20]})
d1=d.apply(add_values, args=(1,2))
print(d1)
```

```python
def multiply(x,m):
    return x*m

d=pd.DataFrame({"A":[1,2],"B":[10,20]})
d1=d.apply(multiply,m=10)
print(d1)
```

## merge                                                                                                                                       
pd.merge(left, right, how="inner or on=None, right_on=None)

```python
e1=pd.DataFrame({
    "emp":["a","b","c"],
    "dep":["it","hr","finance"]
})

e2=pd.DataFrame({
    "emp":["a","b","d"],
    "sal":[50000,40000,30000]
})

r=pd.merge(e1,e2,on="emp",how="inner")
r
```

```python
r1=pd.merge(e1,e2,on="emp",how="outer")
r1
```

```python
r1=pd.merge(e1,e2,on="emp",how="left")
r1
```

```python
r1=pd.merge(e1,e2,on="emp",how="right")
r1
```
