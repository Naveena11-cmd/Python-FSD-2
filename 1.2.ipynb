## Data frame
it is a 2-D array or a table with rows and columns

```python
data={'calories':[400,420,380,390],
      'duration':[50,40,45,60]}
df=pd.DataFrame(data)
df
```

```python
data={'calories':[400,420,380,390],
      'duration':[50,40,45,60]}
df=pd.DataFrame(data)
print(df)
```

```python
# loc displays data on that particular index
df.loc[0]
```

```python
df.iloc[0]
```

```python
data={'calories':[400,420,380,390],
      'duration':[50,40,45,60]}
df=pd.DataFrame(data,index=['day1','day2','day3','day4'])
df
```

- if u change index from default(0,1,2,3) to anything else, iloc will not work but loc will
- and if index is changed then loc will not access from the default index but iloc will
- hence only integer index in iloc
- end index is included in loc but not in iloc

```python
df.loc["day1"]
```

```python
df.iloc["day1"]
```

```python
df[::2]
```

## set_index()
used to set one or more colummns of a datafrane as the index

```python
data={"name":['alice','bob','charlie'],
     "age":[25,30,35],
     "city":["nyc","la","chicago"]}
df=pd.DataFrame(data)
df
```

```python
df.set_index('name', drop=False)
```

```python
df.set_index('name')
```

```python
df
```

```python
df.set_index('name', drop=False, inplace=True)
```

```python
df
```

```python
df.set_index(['name','age'], drop=False)
```

## reset_index()
used to manage and reset the index of a df

```python
data={"name":['alice','bob','charlie'],
     "age":[25,30,35],
     "city":["nyc","la","chicago"]}
f=pd.DataFrame(data, index=['a','b','c'])
f
```

```python
df_reset=f.reset_index()
df_reset
```
