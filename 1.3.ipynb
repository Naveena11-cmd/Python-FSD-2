##  Scatter matrix / pair plot

## parallel_coordinates


```python
from pandas.plotting import parallel_coordinates
d=pd.read_csv('auto-mpg.csv')
parallel_coordinates(d, 'cylinders', cols=['acceleration','mpg','origin'], color=['pink','cyan','tomato','yellow','black'])
```

```python
from pandas.plotting import parallel_coordinates
d=pd.read_csv('auto-mpg.csv')
parallel_coordinates(d, 'displacement', cols=['acceleration','mpg','origin'])
```
