```python
import pandas as pd
```

```python
df=pd.read_csv('tshirt.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Weight']=df['Wight']
```

```python
df.drop('Wight',axis=1, inplace=True)
df.head()
```

```python
x=df[['Weight','Height']]
y=df['Size']
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x,y)
prediction=model.predict(x)
print(prediction)
```

```python
diff=pd.DataFrame({'Actual':y,'Predicted':prediction})
diff
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y, prediction)
print(cm)
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
cm_disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm,
                                       display_labels=['l','m'])
cm_disp.plot()
plt.show()
```

```python
TN=cm[0][0]
FP=cm[0][1]
FN=cm[1][0]
TP=cm[1][1]
```

```python
model_accuracy=(TP+TN)/(TP+FP+FN+TN)
print(model_accuracy)
```

```python
# for what value of K , value of accuracy is most accurate
from sklearn.neighbors import KNeighborsClassifier
acc=[]
for i in range(1,18):
    nn=KNeighborsClassifier(n_neighbors=i)
    model=nn.fit(x,y)
    prediction=model.predict(x)
    cm=confusion_matrix(y,prediction)
    TN=cm[0][0]
    FP=cm[0][1]
    FN=cm[1][0]
    TP=cm[1][1]
    model_accuracy=(TP+TN)/(TP+FP+FN+TN)
    acc.append(model_accuracy)
plt.plot(range(1,18),acc,marker="*")
plt.show()
```

# RANDOM FOREST

**1. Baging (Bootstrap Aggregating)** : *Takes original data and creates random samples with replacement*                                             
**2. Feature Randomness**

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df.isna().sum()
```

```python
df['Sex']=np.where(df['Sex']=='female',0,1)
df.head()
```

```python
x=df[['Pclass','Sex','Age','SibSp','Parch','Fare']]
y=df['Survived']
```

```python
print(x.shape)
print(y.shape)
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier=RandomForestClassifier(criterion='entropy', n_estimators=100,random_state=42)
rf_classifier.fit(x_train,y_train)
pred=rf_classifier.predict(x_test)
print(pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
cm=confusion_matrix(y_test, pred)
print(cm)
```

```python
accuracy_score(y_test, pred)
```

```python
recall_score(y_test, pred)
```

```python
precision_score(y_test, pred)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
choose_tree=rf_classifier.estimators_[0]
plot_tree(choose_tree, feature_names=x.columns, class_names=['Not Survived', 'Survived'], filled=True, rounded=True)
plt.show()
```
