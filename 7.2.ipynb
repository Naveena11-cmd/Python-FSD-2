```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.politifact.com/factchecks/"
result=requests.get(url)
content=result.text
soup=BeautifulSoup(content,'html.parser')
# print(soup.prettify())
```

```python
data=soup.find_all('li',class_='o-listicle__item')
# data
```

```python
data[0]
```

```python
data[0].find('div',class_='m-statement__meta').find('a').text.strip()
```

```python
source=[]
for i in data:
    source.append(i.find('div',class_='m-statement__meta').find('a').text.strip())
print(source)
```

```python
statement=[]
for i in data:
    statement.append(i.find('div',class_='m-statement__quote').text.strip())
print(statement)
```

```python
date=[]
for i in data:
    date.append(i.find('footer',class_='m-statement__footer').text.strip())
print(date)
```

```python
date=[]
for i in data:
    date.append(i.find('footer',class_='m-statement__footer').text.split(' • ')[-1].strip())
print(date)
```

```python
links=[]
root='https://www.politifact.com'
for i in data:
    d=i.find('div',class_='m-statement__quote').find('a')['href'].strip()
    links.append(root+d)
print(links)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Statement':statement, 'Source':source,'Date':date,'Links':links})
df
```

```python
upperframe=[]
for page in range(1,5):
    url='https://www.politifact.com/factchecks/?page'+str(page)
    result=requests.get(url)
    content=result.text
    s=BeautifulSoup(content,'html.parser')
    frame=[]
    data=s.find_all('li',class_='o-listicle__item')
    for i in data:
        source=i.find('div',class_='m-statement__meta').find('a').text.strip()
        statement=i.find('div',class_='m-statement__quote').text.strip()
        date=i.find('footer',class_='m-statement__footer').text.split(' • ')[-1].strip()
        links='https://www.politifact.com'+i.find('div',class_='m-statement__quote').find('a')['href'].strip()
        frame.append((statement,source,date,links))
    upperframe.extend(frame)
print(len(upperframe))
```
